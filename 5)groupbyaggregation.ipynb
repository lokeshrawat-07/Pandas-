```python
import numpy as np
import pandas as pd
```

GroupBy

```python
data = {
    'category' : ['A','B','A','B','A','B','A','B'],
    'store' : ['S1','S2','S1','S2','S2','S1','S2','S1'],
    'sales' : [100,200,400,356,200,400,100,777],
    'quantity' : [2,4,5,6,7,5,4,4],
    'date' : pd.date_range('2025-01-01',periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
cat = df.groupby('category')
cat
```

```python
cat = df.groupby('category')
cat
for i,v in cat:
    print(i)
    print(v)
```

```python
#group by category and calculate their sum
cat = df.groupby('category')['sales'].sum()
cat
```

```python
#group by sales and calculate sum of sales
cat = df.groupby('store')['sales'].sum()
cat
```

```python
#group by multiple column 
#group by category and store
cat = df.groupby(['category','store'])['sales'].sum()
cat
```
