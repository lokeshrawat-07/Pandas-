```python
import numpy as np
import pandas as pd
```

```python
employees = pd.DataFrame({
    'employee_id' : [1,2,3,4,5],
    'name' : ["loki","lucky","rawat","lakku","luke"],
    'deparment': ['HR','IT','Finance','IT','HR']
})


salaries = pd.DataFrame({
    'employee_id' : [1,2,3,6,7],
    'salary' : [17000000,28000000,3800000,4900000,5500000],
    'bonus' : [10000,20000,40000,60000,45000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='employee_id',how='inner')
```

```python
pd.merge(employees,salaries,on='employee_id',how='outer')
```

```python
pd.merge(employees,salaries,on='employee_id',how='left')
```

```python
pd.merge(employees,salaries,on='employee_id',how='right')
```

Concatination of 2 Dataframe

```python
df1 = pd.DataFrame({
    'A' : ['A0','A1','A2'],
    'B' : ['B0','B1','B2'],
    'C' : ['C0','C1','C2']
})
df2 = pd.DataFrame({
    'A' : ['A3','A4','A5'],
    'B' : ['B3','B4','B5'],
    'C' : ['C3','C4','C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

joining of two dataframe

```python
df1 = pd.DataFrame({
    'name' : ["bob","alice","jone"]
}, index = [1,2,3])
df2 = pd.DataFrame({
    'score' : [90,89,98]
}, index = [2,3,4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2,how='outer')
```
