make a new column for episode count

```python
df.loc[0]['Title']
```

```python
def extract_episode(txt):
    data = ""
    check = False
    for i in txt:
        if i == ")":
            check = False
            return data
        if check == True:
            data += i
        if i == '(':
            check = True
```

```python
df['episode'] = df["Title"].apply(extract_episode)
```

```python
df['episode'] = df['episode'].str.replace(" episodes","")
```

```python
type(df.loc[0]['episode'])
```

```python
df['episode'] = df['episode'].astype('int')
```

make a new column for time stamp

```python
df.loc[3]['Score']
```

```python
def extraction_score(txt):
    data2 = ""
    check = False
    for i in txt:
        if i == "!":
            return data2
        if check == True:
            data2 += i
        if i == ".":
            check = True
            
       
    return data2
```

```python
df['new_score'] = df['Score'].apply(extraction_score)
```

```python
# which anime has the highest newscore 
df[df['new_score'] == df['new_score'].max()]['Title']
```

```python
#which anime has the highest episode
df[df['episode'] == df['episode'].max()]
```
